```python
#I want to start with using analyzing the numerical variables first with boxplot and histogram at the same time 
def histogram_boxplot(feature, figsize=(15,10), bins = None):
    """ Boxplot and histogram combined
    feature: 1-d feature array
    figsize: size of fig (default (9,8))
    bins: number of bins (default None / auto)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(nrows = 2, # Number of rows of the subplot grid= 2
                                           sharex = True, # x-axis will be shared among all subplots
                                           gridspec_kw = {"height_ratios": (.25, .75)}, 
                                           figsize = figsize 
                                           ) # creating the 2 subplots
    sns.boxplot(feature, ax=ax_box2, showmeans=True, color='yellow') # boxplot will be created and a triangle will indicate the mean value of the column
    sns.distplot(feature, kde=F, ax=ax_hist2, bins=bins) if bins else sns.distplot(feature, kde=False, ax=ax_hist2) # For histogram
    ax_hist2.axvline(np.mean(feature), color='g', linestyle='-') # Add mean to the histogram
    ax_hist2.axvline(np.median(feature), color='black', linestyle='--') # Add median to the histogram
```

Lets move onto Multivariate analysis of our dataset

To start off, we use multivariate analysis to understand interactions between different fields in the dataset or finding interactions between variables more than two. 

Above, in the statistical analysis of this dataset we used the correlation function to dislay the correlation between every possible pair of attributes as a dataframe. However, here it is again for your analysis. 

```python
data.corr() #this is the correlation function 
```

```python
#we can use representation from the correlation function graphically by using the heatmap function in seaborn 
num_var = ['Age','Education','Usage','Fitness','Income', 'Miles']

corr = data[num_var].corr()

# plot the heatmap
plt.figure(figsize=(14,10))
sns.heatmap(corr, annot=True,cmap='coolwarm',
        fmt=".1f",
        xticklabels=corr.columns,
        yticklabels=corr.columns); #by adding the column at the end we don't see any extra text at the top
```

By using the heatmap, we are able to see which two variables strongly correlate with eachother. Lighter correlations are higher correlations and closer to one. We see that Miles and Usage are strongly correlated, which is no surprise. Miles and fitness are also strongly correlated, which is no surprise. Fitness and usage comes shortly behind. We see there is little correlation between Miles and Age, Usage and Age, as well as Fitness and Age. 

Another way to see a two dimensional data set is though the jointplot function. This will create a scatterplot of two variables along with the histograms of each next to the appropriate axes. 

```python
#lets use the jointplot function on our strongly correlated variables 
sns.jointplot(data['Miles'], data['Fitness']);
```

```python
#another strongly correlated variable pairis Miles and Usage
sns.jointplot(data['Miles'], data['Usage']);
```

```python
#fitness and usage were a frontrunner on the strongly correlated as well  

sns.jointplot(data['Usage'], data['Fitness']);
```

From the three jointplots we get to see that those with a fitness score close to 1, choose to run less miles. As the fitness score increases, we see that the miles increase as well. 

While there are some outliers on the Usage and Miles jointplot, we see for the most part that if the person plans on using the treadmill more days, that there is a higher chance of them going more miles. 

Between Fitness and Usage, you see that the usage is spread apart.
